## BREAST CANCER WISCONSIN (DIAGNOSTIC)
- **Description**: Features computed from a digitized image of a fine needle aspirate (FNA) of a breast mass describe characteristics of the cell nuclei present in the image.
- **Target Variable Type**: Binary (e.g., malignant or benign).
- **Problem Statement**: Develop a machine learning model that analyzes the characteristics of cell nuclei from digitized images of breast mass FNA to accurately distinguish between benign and malignant tumors.
- **Most Important Performance Metric**: Precision and Recall (F1 Score).
    - **Justification**: For medical diagnostic problems, both false positives and false negatives carry significant consequences. Precision is critical because false positives can lead to unnecessary anxiety, stress, and further invasive testing for patients. Recall is equally important because missing a malignant tumor (false negative) can have life-threatening implications.
    - **Real-Life Scenario**: In a screening program for breast cancer, a high F1 score ensures that the majority of malignancies are correctly identified (high recall) while minimizing the stress and cost of false alarms (high precision).
        - **Note**: The F1 Score is the harmonic mean of precision and recall, providing a balanced evaluation metric for binary classification tasks.
